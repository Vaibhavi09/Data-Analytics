This project explores diabetes prediction using a dataset with features like glucose levels, BMI, and age. Utilizing logistic regression, we achieved accurate predictions. The analysis includes a visually appealing table and insightful visualizations, offering a concise yet informative overview of the data exploration and modeling process.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from tabulate import tabulate
```

```python
# Load the dataset
url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
gapminder = pd.read_csv(url)
```

```python
# Display basic information about the dataset
info = gapminder.info()
```

```python
# Display the first few rows of the dataset
head = gapminder.head()
```

```python
# Display Dataset Information in a Table
info_table = tabulate(pd.DataFrame(info).transpose(), headers='keys', tablefmt='pretty', showindex=False)

# Display First Few Rows of the Dataset in a Table
head_table = tabulate(head, headers='keys', tablefmt='pretty', showindex=False)

# Print the tables
print("Dataset Information:")
print(info_table)

print("\nFirst Few Rows of the Dataset:")
print(head_table)
```

```python
# Filter data for a specific year (taking 2007)
year_to_analyze = 2007
data_2007 = gapminder[gapminder['year'] == year_to_analyze]
```

```python
# Scatter plot: Life Expectancy vs GDP per Capita
plt.figure(figsize=(12, 8))
plt.scatter(data_2007['gdpPercap'], data_2007['lifeExp'], s=data_2007['pop'] / 1e6, alpha=0.7, cmap='viridis')
plt.title(f'Scatter Plot: Life Expectancy vs GDP per Capita ({year_to_analyze})')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.colorbar(label='Population (millions)')
plt.show()
```

```python
# Line plot: Life Expectancy and GDP per Capita trend over the years for a specific country (taking - United States)
country_to_analyze = 'United States'
data_country = gapminder[gapminder['country'] == country_to_analyze]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data_country['year'], data_country['lifeExp'], label='Life Expectancy', marker='o')
plt.plot(data_country['year'], data_country['gdpPercap'], label='GDP per Capita', marker='o')
plt.title(f'Trend of Life Expectancy and GDP per Capita over the Years ({country_to_analyze})')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Line plot: Life Expectancy and GDP per Capita trend over the years for a specific country (taking India)
country_to_analyze = 'India'
data_country = gapminder[gapminder['country'] == country_to_analyze]


plt.figure(figsize=(12, 6))
plt.plot(data_country['year'], data_country['lifeExp'], label='Life Expectancy', marker='o')
plt.plot(data_country['year'], data_country['gdpPercap'], label='GDP per Capita', marker='o')
plt.title(f'Trend of Life Expectancy and GDP per Capita over the Years ({country_to_analyze})')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data_2007 = gapminder[gapminder['year'] == 2007]
plt.figure(figsize=(10, 6))
sns.barplot(x='continent', y='lifeExp', data=data_2007, ci=None, palette='viridis')
plt.title('Average Life Expectancy Across Continents in 2007')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy')
plt.show()
```
