Project Name - Diamonds
Here the data is being derived and mounted by a github url. This analysis is created using matplolib library in python. This dataset describes the different variables on with the data relies. Insightful visualisation have be created using python only.

```python
# Import the libraries (that will be used)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

# Reading the dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv'
diamonds_df = pd.read_csv(url)

# Overview of the dataset
diamonds_df.head()
```

```python
# Summary statistics of the dataset
diamonds_df.describe()
```

```python
# Distribution of diamond prices (thin bar graph)
sns.histplot(data=diamonds_df, x='price', kde=True)
plt.show()
```

```python
# Scattered plotting of carat vs price
sns.scatterplot(data=diamonds_df, x='carat', y='price')
plt.show()
```

```python
# Bar graph plotting of price by cut
sns.barplot(data=diamonds_df, x='cut', y='price')
plt.show()
```
